```python
def read_txt(path):
    with open(path) as file:
        text = file.read()
    file.close()
    return(text)
```

```python
def chr_to_bin(text):
    i=0
    s=''
    while i<len(text):
        tmp_ord=ord(text[i])
        tmp_s='{0:b}'.format(tmp_ord)
        while len(tmp_s)%11!=0:
            tmp_s='0'+tmp_s
        s+=tmp_s
        i+=1
    while len(s)%64!=0:
        s='0'+s
    return(s)
```

```python
def func_s(bp,s):
    strg=int(bp[0]+bp[5],2)
    clmn=int(bp[1:5],2)
    res=s[strg*16+clmn]
    res='{0:b}'.format(res)
    res='{:0>4}'.format(res)
    return(res)
```

```python
def iround(a,b,k):
    ai=b
    f_res=func_f(b,k)
    i=0
    bi=''
    while i<32:
        if a[i]==f_res[i]:
            bi+='0'
        else:
            bi+='1'
        i+=1
    return(ai,bi)
```
